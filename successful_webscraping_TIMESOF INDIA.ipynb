# Calling URL and getting contents from it

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
url="https://timesofindia.indiatimes.com/entertainment/tamil/movie-reviews"
result=requests.get(url)
webpage=result.content
sl_soup=BeautifulSoup(webpage,"html.parser")
sl_soup
```

# For loop to view all the tags present in content

```python

for tag in sl_soup.find_all(True):
    print(tag.name)
result.close
```

# To jump in to this part you have to find the class under which your data is lying 

```python
entire_table=sl_soup.findAll('div',attrs={'class':'FIL_right'})
entire_table
```

# Under the Class we are taking one part and trying to extract the required data

```python
entire_table[0]
```

# (small>Tamil<br/>Crime, Mystery, Thriller | UA</small) inside this i am able to find all the contents i require so iwill split this to array to get data

```python
entire_table[0].small.text
language=entire_table[0].small.text.split()
print("language=",language[0])
print("Genre=",language[1])
print("Certification=",language[4])
```

# After confirming we will run for loop to get all the required data through out the table

```python
movie_title=[]
GENRE=[]
CERTIFICATION=[]
for movie_name in entire_table:
    name=movie_name.h3.text
    movie_title.append(name)
for DATA in entire_table:
    SP_DATA = DATA.small.text.split('|')
    GENRE.append(SP_DATA[0]) 
    try:
        CERTIFICATION.append(SP_DATA[1])
    except:
        CERTIFICATION.append("not available")
movietitle=pd.DataFrame({'MOVIE NAME':movie_title})
df_genre=pd.DataFrame({'GENRE':GENRE})
data_df=pd.concat([movietitle,df_genre],axis=1,join='inner')
df_cert=pd.DataFrame(CERTIFICATION)
data_df['CERTIFICATE']=df_cert
data_df
```
